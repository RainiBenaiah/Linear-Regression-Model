```python
#Handling the missing values
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
# initialiing imtuter
imputer = IterativeImputer(random_state=46, verbose=True)

imputed_data = imputer.fit_transform(data)
data_impute = pd.DataFrame(data, columns = data.columns)
data.head()
data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
#dropping some columns
partial_data_cols = ['Air temperature (C)', 'Wind speed (Km/h)',
                     'Air humidity (%)','Wind gust (Km/h)','Pressure (KPa)']
data[partial_data_cols] = data[partial_data_cols].fillna(data[partial_data_cols].mean())

sparse_cols = ['ph', 'rainfall', 'N', 'P', 'K','Status']
# Check if the columns exist before dropping
sparse_cols_to_drop = [col for col in sparse_cols if col in data.columns]

# Drop only the existing columns
new_data = data.drop(columns=sparse_cols_to_drop)

numeric_cols = ['Soil Moisture', 'Temperature', ' Soil Humidity', 'Time'] + partial_data_cols
scaler = StandardScaler()
new_data[numeric_cols] = scaler.fit_transform(data[numeric_cols])

new_data.head()
```

```python
# Standardizing the  Data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
new_data = pd.DataFrame(scaler.fit_transform(new_data), columns=new_data.columns)
new_data.head()
```

```python
#Training the Linear Model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = new_data.drop('Soil Moisture', axis=1)
y = new_data['Soil Moisture']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# Strip leading/trailing spaces from all column names in both training and testing data
x_train.columns = x_train.columns.str.strip()
x_test.columns = x_test.columns.str.strip()

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# Root Mean S quareroot
from sklearn.metrics import mean_squared_error

y_pred = lr.predict(x_test)
mse_lr = mean_squared_error(y_test, y_pred)
rmse_lr = np.sqrt(mse_lr)

print("Linear Regression RMSE:", rmse_lr)
```

```python
# Comparing the RMSE of Linear Regression with other models
#   Random Forest and Desision Trees and compare RSMES
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
# random forest
random_forest = RandomForestRegressor(random_state=42)
random_forest.fit(x_train, y_train)
Y_pred_rf = random_forest.predict(x_test)
rmse_rf = np.sqrt(mean_squared_error(y_test, Y_pred_rf))

# decision tree
decision_tree = DecisionTreeRegressor(random_state=42)
decision_tree.fit(x_train, y_train)
Y_pred_dt = decision_tree.predict(x_test)
rmse_dt = np.sqrt(mean_squared_error(y_test, Y_pred_dt))
print("Random Forest RMSE:", rmse_rf)
print("Decision Tree RMSE:", rmse_dt)
model_rank = {
     'linear regression' : rmse_lr,
     'Random Forest' : rmse_dt,
     'Decision Trees' : rmse_rf


}
print (model_rank)
```

```python
sorted_model_rank = dict(sorted(model_rank.items(), key=lambda item: item[1]))
print("Model Rankings based on RMSE:")
for model, rmse in sorted_model_rank.items():
    print(f"{model}: RMSE = {rmse}")
```
