```python
from sklearn.preprocessing import StandardScaler
#dropping some columns
partial_data_cols = ['Air temperature (C)', 'Wind speed (Km/h)',
                     'Air humidity (%)','Wind gust (Km/h)','Pressure (KPa)']
data[partial_data_cols] = data[partial_data_cols].fillna(data[partial_data_cols].mean())

sparse_cols = ['ph', 'rainfall', 'N', 'P', 'K']
# Check if the columns exist before dropping
sparse_cols_to_drop = [col for col in sparse_cols if col in data.columns]

# Drop only the existing columns
data = data.drop(columns=sparse_cols_to_drop)

numeric_cols = ['Soil Moisture', 'Temperature', ' Soil Humidity', 'Time'] + partial_data_cols
scaler = StandardScaler()
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])

data.head()
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.impute import SimpleImputer
# Assuming 'data' is your DataFrame and 'Soil Moisture' is the target variable
Y = data['Soil Moisture']  # Defining Y explicitly

X = data.drop(['Soil Moisture'], axis=1)  # Original X as DataFrame
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X) # Imputed data as array

# Convert the imputed array back to a DataFrame with original column names
X = pd.DataFrame(X_imputed, columns=X.columns, index=data.index) # X is now DataFrame again, with original index

vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
print(vif_data)
features_to_drop = vif_data[vif_data['VIF'] > 10]['feature']
X = X.drop(features_to_drop, axis=1)

# Align Y with the potentially modified X index
Y = Y[X.index] #Ensuring Y's index matches X's index

# Build your model
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()

print(model.summary())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Handling missing data for columns with partial data
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
model = LinearRegression()
# train your model
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python


rmse = np.sqrt(mse)
print("Root Mean Square Error:", rmse)
```

```python
#  estimator object for Random Forest and Desision Trees and compare RSMES
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
# random forest
random_forest = RandomForestRegressor(random_state=42)
random_forest.fit(X_train, Y_train)
Y_pred_rf = random_forest.predict(X_test)
rmse_rf = np.sqrt(mean_squared_error(Y_test, Y_pred_rf))

# decision tree
decision_tree = DecisionTreeRegressor(random_state=42)
decision_tree.fit(X_train, Y_train)
Y_pred_dt = decision_tree.predict(X_test)
rmse_dt = np.sqrt(mean_squared_error(Y_test, Y_pred_dt))
print("Random Forest RMSE:", rmse_rf)
print("Decision Tree RMSE:", rmse_dt)
model_rank = {
     'linear regression' : rmse,
     'Random Forest' : rmse_dt,
     'Decision Trees' : rmse_rf


}
print (model_rank)
```

```python
sorted_model_rank = dict(sorted(model_rank.items(), key=lambda item: item[1]))
print("Model Rankings based on RMSE:")
for model, rmse in sorted_model_rank.items():
    print(f"{model}: RMSE = {rmse}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

param_grid = {
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 2, 5],
    'max_features': [None, 'sqrt', 'log2']
}


grid_search = GridSearchCV(estimator=DecisionTreeRegressor(random_state=42),
                           param_grid=param_grid,
                           scoring='neg_mean_squared_error',
                           cv=5, verbose=2, n_jobs=-1)

grid_search.fit(X_train, Y_train)

# Best parameters and model
print("Best Parameters:", grid_search.best_params_)
best_rf_model = grid_search.best_estimator_

```

```python
# Train Decision Tree with best parameters
best_dt_model = DecisionTreeRegressor(
    max_depth=9,
    max_features=None,
    min_samples_leaf=5,
    min_samples_split=20,
    random_state=42
)

best_dt_model.fit(X_train, Y_train)

# Make predictions
y_pred_best = best_dt_model.predict(X_test)

# Calculate RMSE
from sklearn.metrics import mean_squared_error
rmse_best = mean_squared_error(Y_test, y_pred_best, squared=False)
print(f"Best Decision Tree RMSE: {rmse_best}")
```

```python
residuals = Y_test - y_pred_best
plt.scatter(y_pred_best, residuals, alpha=0.6)
plt.axhline(0, color='r', linestyle='--', linewidth=1)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Predicted Values (Tuned Random Forest)")
plt.show()
```

```python
# Load the saved model
with open(model_filename, 'rb') as file:
    loaded_model = pickle.load(file)

print("Model loaded successfully!")

```
